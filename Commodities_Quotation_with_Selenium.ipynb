Data base: https://drive.google.com/drive/folders/1O9aE_Hen5smf_a6TsbVF6DM0ms4Ykj6s?usp=share_link  - "commodities.xlsx", this file needs to be in the same folder of the one that we are writing the code

- Do not forget to download the webdriver

```python
#step by step

# step 1 - open the browser

from selenium import webdriver

browser = webdriver.Chrome()
browser.get("https://www.google.com/")
```

```python
# step 2 - import data base
import pandas as pd

table = pd.read_excel('commodities.xlsx')
display(table)
```

```python
for line in table.index:
    product = table.loc[line, "Produto"]
    print(product)
    product = product.replace('ó','o').replace('ã','a').replace('á','a').replace('ç','c').replace('ú','u').replace('é','e')
    
    link = f'https://www.melhorcambio.com/{product}-hoje'
    browser.get(link)

    price = browser.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    price = price.replace('.','').replace(',','.')
    print(price)
    table.loc[line, 'Preço Atual'] = float(price)
    
display(table)
# step 3 - for each commodity in data base:
        # step 4 - search commodity price
        # step 5 - update prices in data base
        # step 6 - choose which commodities to buy
```

```python
# fill Comprar Column
table['Comprar'] = table['Preço Atual'] < table['Preço Ideal']
display(table)

# export data base to excel
table.to_excel('commodities_update.xlsx', index= False )
browser.quit()
```
